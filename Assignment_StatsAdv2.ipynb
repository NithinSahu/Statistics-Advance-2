# Statistics Part 2 (Theory Questions)

###1.  What is hypothesis testing in statistics?
- Hypothesis testing is a statistical method used to make inferences or draw conclusions about a population based on sample data.
-  It allows analysts to test assumptions (hypotheses) regarding population parameters, such as means or proportions, by evaluating whether observed data provide enough evidence to reject a specific claim (the null hypothesis).

###2. What is the null hypothesis, and how does it differ from the alternative hypothesis?
- The null hypothesis states that nothing significant is happening, while the alternative hypothesis suggests that something significant is present.

| Parameter              | Null Hypothesis (H₀)                                                      | Alternative Hypothesis (H₁ or Ha)                                    |
|------------------------|----------------------------------------------------------------------------|-----------------------------------------------------------------------|
| Definition             | A claim that there is no effect or relationship in the population. | A claim that there is an effect or relationship in the population. |
| What is it?            | Generally, researchers try to reject or disprove it.                 | Researchers try to accept or prove it.                         |
| Testing Process        | Indirect and implicit.                                                  | Direct and explicit.                                              |
| Notation               | H₀                                                                         | H₁ or Ha                                                             |
| Symbol Used            | Equality symbol (=, ≥, ≤).                                     | Inequality symbol (≠, ).                             |
| Result                 | Indicates no change in opinion or action.                               | Causes a change in opinion or action if supported.                |
| Acceptance Criteria    | Accepted if p-value > alpha (α).                                  | Accepted if p-value < alpha (α).                            |

###3. What is the significance level in hypothesis testing, and why is it important?
**Significance Level in Hypothesis Testing (α):**

- The significance level (alpha, α) is the threshold for deciding whether a result is statistically significant, commonly set at 0.05 (5%).
- It represents the probability of making a Type I error—rejecting the null hypothesis when it is actually true.
- If the p-value is less than or equal to α, you reject the null hypothesis.
- The choice of α is important because it controls the risk of false positives and should reflect the seriousness of potential errors in your context.

**Example Table:**

| α Value | Meaning                                 |
|---------|-----------------------------------------|
| 0.05    | 5% risk of Type I error (common)        |
| 0.01    | 1% risk (stricter, more cautious)       |


###4. What does a P-value represent in hypothesis testing?
- A p-value shows the probability of getting your observed results (or more extreme) if the null hypothesis is true. A small p-value means your results are unlikely due to chance and may be statistically significant.

###5. How do you interpret the P-value in hypothesis testing?
- If p-value ≤ α: reject the null hypothesis.
- If p-value > α: do not reject the null hypothesis.

###6. What are Type 1 and Type 2 errors in hypothesis testing?
- Type 1 (False Positive): Rejecting a true null hypothesis (e.g., claiming a drug works when it doesn’t). Probability = α (e.g., 0.05) .
- Type 2 (False Negative): Failing to reject a false null hypothesis (e.g., missing a drug’s effect). Probability = β .

###7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
- One-tailed: Tests for an effect in one direction (e.g., "Group A is greater than Group B").
- Two-tailed: Tests for any difference (e.g., "Group A ≠ Group B").

###8. What is the Z-test, and when is it used in hypothesis testing?
- Z-test Compares sample mean to population mean when σ (population SD) is known and sample size is large (n ≥ 30).
- Example: Testing if a class’s average score differs from the national average with known σ .

###9. How do you calculate the Z-score, and what does it represent in hypothesis testing?
- Z-score
  - Z = (Sample Mean − Population Mean) / (σ / sqrt(n))
- Represents how many SDs a sample mean is from the population mean.

###10.  What is the T-distribution, and when should it be used instead of the normal distribution?
- T-distribution vs. Normal Distribution:
  - T-distribution: Flatter and wider tails for small samples (n < 30) or unknown σ.
  - Used in T-tests (e.g., comparing two small sample means).

###11. What is the difference between a Z-test and a T-test?
- Z-test: Large samples, known σ.
- T-test: Small samples, unknown σ. Both compare means.

###12. What is the T-test, and how is it used in hypothesis testing?
T-test Applications
- One-sample: Compare sample mean to a known value.
- Independent samples: Compare two groups.
- Paired: Compare pre/post measurements in the same group.

###13. What is the relationship between Z-test and T-test in hypothesis testing?
Z-test vs. T-test Relationship
- Both assess mean differences; T-test adjusts for uncertainty in small samples/unknown σ.

###14. What is a confidence interval, and how is it used to interpret statistical results?
Confidence Interval (CI)
- A range (e.g., 95% CI) where the true population parameter likely lies.
- Interpretation: "We’re 95% confident the mean is between X and Y".

###15. What is the margin of error, and how does it affect the confidence interval?
Margin of Error
- Half the CI width. Affected by sample size (larger n → smaller margin) and variability.

###16. How is Bayes' Theorem used in statistics, and what is its significance?
Bayes’ Theorem
- Updates probabilities using prior knowledge + new data. Key in Bayesian statistics (e.g., spam filters updating email classification).

###17. What is the Chi-square distribution, and when is it used?
Chi-square Distribution
- Used for categorical data tests (e.g., goodness-of-fit, independence). Asymmetrical and right-skewed.

###18. What is the Chi-square goodness of fit test, and how is it applied?
Chi-square Goodness-of-Fit Test
- Checks if observed categorical data matches expected distribution (e.g., testing if dice are fair).

###19. What is the F-distribution, and when is it used in hypothesis testing?
F-distribution
- Used in ANOVA to compare variances. Right-skewed.
- F-test: Compares model variances (e.g., ANOVA, regression).

###20. What is an ANOVA test, and what are its assumptions?
ANOVA Test
- Compares ≥3 group means. Assumptions:
Normality.
Equal variances.
Independent observations.

###21. What are the different types of ANOVA tests?
Types: One-way (1 factor), Two-way (2 factors).

###22. What is the F-test, and how does it relate to hypothesis testing?
F-test in ANOVA
- Ratio of between-group variance to within-group variance.
- High F-value → significant group differences.

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.

def bernoulli_pdf(k, p):
    """
    Calculate the probability distribution function (PDF/PMF) of a Bernoulli distribution.

    Parameters:
        k (int): The outcome, should be 0 or 1.
        p (float): Probability of success (1), 0 <= p <= 1.

    Returns:
        float: Probability of observing outcome k.
    """
    if k not in [0, 1]:
        raise ValueError("k must be 0 or 1 for Bernoulli distribution")
    return p**k * (1 - p)**(1 - k)

# Example usage:
print("P(X=1):", bernoulli_pdf(1, 0.3))  # Output: 0.3
print("P(X=0):", bernoulli_pdf(0, 0.3))  # Output: 0.7
```

```python
#9.  Write a Python function to calculate Z-scores from a dataset and plot them.

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_and_plot_zscores(data):
    """
    Calculate Z-scores for a dataset and plot them.
    Args:
        data (array-like): Input dataset (list or numpy array)
    """
    # Calculate Z-scores using scipy
    z_scores = stats.zscore(data)  # [2][3][4][5]

    # Plot Z-scores
    plt.plot(z_scores, marker='o', linestyle='-', color='b')
    plt.axhline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.xlabel('Data Index')
    plt.ylabel('Z-score')
    plt.title('Z-scores of the Dataset')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
    return z_scores

# Example usage:
data = np.array([12, 15, 18, 21, 24, 27, 30, 33, 36, 39])
calculate_and_plot_zscores(data)
```

```python
#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

import numpy as np
from scipy.stats import norm

# Example dataset and data point
data = np.array([12, 15, 18, 21, 24, 27, 30, 33, 36, 39])
x = 27  # Data point to calculate Z-score for

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Calculate Z-score
z_score = (x - mean) / std_dev
print(f"Z-score for {x}: {z_score:.3f}")

# Compare to standard normal distribution
prob = norm.cdf(z_score)  # Probability that a value is less than or equal to x
print(f"Probability under standard normal curve up to Z={z_score:.3f}: {prob:.3f}")

# Optional: Visualize
import matplotlib.pyplot as plt

x_vals = np.linspace(-4, 4, 1000)
y_vals = norm.pdf(x_vals)

plt.plot(x_vals, y_vals, label='Standard Normal PDF')
plt.fill_between(x_vals, 0, y_vals, where=(x_vals <= z_score), color='skyblue', alpha=0.5, label=f'P(Z ≤ {z_score:.2f})')
plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.title('Standard Normal Distribution and Z-score')
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
#15. Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
from scipy.stats import norm

# Sample data and population parameters
sample = np.array([110, 108, 115, 120, 112, 117, 109, 113, 116, 114])  # Example sample
sample_mean = np.mean(sample)
n = len(sample)
population_mean = 100  # μ
population_std = 15    # σ
alpha = 0.05           # Significance level

# Step 1: State hypotheses
# H0: Sample mean = Population mean
# H1: Sample mean ≠ Population mean

# Step 2: Calculate standard error
standard_error = population_std / np.sqrt(n)

# Step 3: Calculate Z-statistic
z_stat = (sample_mean - population_mean) / standard_error

# Step 4: Calculate p-value (two-tailed)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print(f"Sample mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Step 5: Decision
if p_value < alpha:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

```python
#24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution

import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    """
    Calculates the skewness and kurtosis of a dataset.

    Args:
        data (array-like): The input dataset.

    Returns:
        tuple: A tuple containing the skewness and kurtosis values.
    """
    # Calculate skewness
    data_skewness = skew(data)
    # Calculate kurtosis (Fisher's definition, excess kurtosis)
    data_kurtosis = kurtosis(data)
    return data_skewness, data_kurtosis

# Generate a sample dataset from a normal distribution
mu = 0      # Mean
sigma = 1   # Standard deviation
size = 1000 # Number of data points
normal_data = np.random.normal(loc=mu, scale=sigma, size=size)

# Calculate and print skewness and kurtosis
skewness_val, kurtosis_val = calculate_skewness_kurtosis(normal_data)

print(f"Skewness of the normal dataset: {skewness_val:.4f}")
print(f"Kurtosis of the normal dataset: {kurtosis_val:.4f}")

# Example with a non-normal distribution (e.g., exponential)
exponential_data = np.random.exponential(scale=1, size=1000)
skewness_exp, kurtosis_exp = calculate_skewness_kurtosis(exponential_data)

print(f"\nSkewness of the exponential dataset: {skewness_exp:.4f}")
print(f"Kurtosis of the exponential dataset: {kurtosis_exp:.4f}")
```

```python
#3. Implement a one-sample Z-test using Python to compare the sample mean with the population mean

import numpy as np
def z_test_one_sample(sample_mean, pop_mean, pop_std_dev, sample_size, alpha=0.05, alternative='two-sided'):
    """
    Performs a one-sample Z-test to compare a sample mean with a population mean.

    Args:
        sample_mean (float): The mean of the sample.
        pop_mean (float): The known mean of the population (μ0).
        pop_std_dev (float): The known standard deviation of the population (σ).
        sample_size (int): The size of the sample (n).
        alpha (float, optional): The significance level (default is 0.05).
        alternative (str, optional): Specifies the alternative hypothesis.
                           'two-sided' (default): H1: sample mean != pop mean
                           'greater': H1: sample mean > pop mean
                           'less': H1: sample mean < pop mean

    Returns:
        tuple: A tuple containing:
            - z_score (float): The calculated Z-score.
            - p_value (float): The calculated p-value.
            - decision (str): The conclusion based on the Z-test.
    """
    # Calculate the standard error of the mean
    standard_error = pop_std_dev / np.sqrt(sample_size)

    # Calculate the Z-score
    z_score = (sample_mean - pop_mean) / standard_error

    # Calculate the p-value based on the alternative hypothesis
    if alternative == 'two-sided':
        p_value = 2 * (1 - norm.cdf(abs(z_score)))
    elif alternative == 'greater':
        p_value = 1 - norm.cdf(z_score)
    elif alternative == 'less':
        p_value = norm.cdf(z_score)
    else:
        raise ValueError("Invalid alternative. Choose 'two-sided', 'greater', or 'less'.")

    # Make a decision
    if p_value < alpha:
        decision = f"Reject H0 at alpha={alpha} (p-value={p_value:.4f} < {alpha})"
    else:
        decision = f"Fail to reject H0 at alpha={alpha} (p-value={p_value:.4f} >= {alpha})"

    return z_score, p_value, decision

# --- Example Usage ---

# Scenario: A company claims the average response time for their service is 30 minutes.
# The known population standard deviation of response times is 5 minutes.
# A sample of 40 recent service calls has an average response time of 32 minutes.
# We want to test if the sample mean is significantly different from the claimed population mean at a 5% significance level.

# Define parameters
sample_mean_obs = 32.0      # Observed sample mean
population_mean_h0 = 30.0   # Hypothesized population mean under H0 (μ0)
population_std_known = 5.0  # Known population standard deviation (σ)
sample_size_n = 40          # Sample size (n)
alpha_level = 0.05          # Significance level

# Perform the one-sample Z-test (two-sided test)
z_stat, p_val, test_decision = z_test_one_sample(
    sample_mean_obs,
    population_mean_h0,
    population_std_known,
    sample_size_n,
    alpha=alpha_level,
    alternative='two-sided'
)

# Print the results
print("--- One-Sample Z-Test Results ---")
print(f"Sample Mean: {sample_mean_obs}")
print(f"Hypothesized Population Mean (H0): {population_mean_h0}")
print(f"Known Population Std Dev: {population_std_known}")
print(f"Sample Size: {sample_size_n}")
print(f"Significance Level (alpha): {alpha_level}")
print(f"Alternative Hypothesis: Two-sided (mean != {population_mean_h0})")
print("-" * 30)
print(f"Calculated Z-score: {z_stat:.4f}")
print(f"Calculated P-value: {p_val:.4f}")
print("-" * 30)
print(f"Decision: {test_decision}")

# Interpretation
if 'Reject' in test_decision:
    print(f"\nInterpretation: Based on the Z-test, there is sufficient evidence to reject the null hypothesis.")
    print(f"This suggests that the true average response time is significantly different from {population_mean_h0} minutes.")
else:
    print(f"\nInterpretation: Based on the Z-test, we fail to reject the null hypothesis.")
    print(f"There is not enough evidence to conclude that the true average response time is significantly different from {population_mean_h0} minutes.")
    print("The observed difference could be due to random sampling variability.")
```

```python
#4.  Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import matplotlib.pyplot as plt
import numpy as np
def plot_z_test_decision_region(population_mean, population_std, sample_size, alpha=0.05):
    """
    Visualizes the decision regions for a two-tailed Z-test on a standard normal distribution.

    Args:
        population_mean (float): The known mean of the population (μ0).
        population_std (float): The known standard deviation of the population (σ).
        sample_size (int): The size of the sample (n).
        alpha (float, optional): The significance level (default is 0.05) for a two-tailed test.
    """
    # Generate x values for the standard normal distribution
    x_vals = np.linspace(-4, 4, 1000)
    # Calculate the PDF for the standard normal distribution
    pdf_vals = norm.pdf(x_vals)

    # Calculate the critical Z-values for a two-tailed test
    # For a two-tailed test, alpha is split into two tails (alpha/2 on each side)
    z_critical_upper = norm.ppf(1 - alpha / 2)
    z_critical_lower = norm.ppf(alpha / 2) # This will be the negative of z_critical_upper

    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, pdf_vals, label='Standard Normal Distribution')

    # Shade the rejection regions
    plt.fill_between(x_vals, 0, pdf_vals, where=(x_vals >= z_critical_upper), color='red', alpha=0.3, label=f'Rejection Region (α/2 = {alpha/2})')
    plt.fill_between(x_vals, 0, pdf_vals, where=(x_vals <= z_critical_lower), color='red', alpha=0.3)

    # Shade the acceptance region
    plt.fill_between(x_vals, 0, pdf_vals, where=((x_vals > z_critical_lower) & (x_vals < z_critical_upper)), color='green', alpha=0.2, label='Acceptance Region (1 - α)')

    plt.axvline(z_critical_upper, color='red', linestyle='--', linewidth=1.5, label=f'Critical Z = {z_critical_upper:.2f}')
    plt.axvline(z_critical_lower, color='red', linestyle='--', linewidth=1.5)
    plt.axvline(0, color='grey', linestyle='-', linewidth=0.8) # Line at Z=0

    plt.title(f'Two-Tailed Z-Test Decision Regions (α = {alpha})')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example Usage for plotting the decision region:
# We use the parameters from the previous Z-test example for context,
# but the plot itself only shows the standard normal distribution and critical values,
# not a specific sample result.
population_mean_h0 = 30.0   # Hypothesized population mean under H0 (μ0)
population_std_known = 5.0  # Known population standard deviation (σ)
sample_size_n = 40          # Sample size (n)
alpha_level = 0.05          # Significance level

# Plot the decision region for the specified alpha
plot_z_test_decision_region(population_mean_h0, population_std_known, sample_size_n, alpha=alpha_level)

# To demonstrate how a sample Z-score relates to the region,
# let's use the Z-statistic calculated in the previous example (z_stat)
# and overlay it on the plot.

# Assuming the z_stat from the previous example is available
# If not, uncomment the previous Z-test calculation block or run it first.
# z_stat was calculated as (sample_mean_obs - population_mean_h0) / (population_std_known / np.sqrt(sample_size_n))
# Using sample_mean_obs = 32, pop_mean=30, pop_std=5, n=40:
z_stat_example = (32 - 30) / (5 / np.sqrt(40))
print(f"\nExample Sample Z-statistic: {z_stat_example:.4f}")


def plot_z_test_with_sample(sample_mean, population_mean, population_std, sample_size, alpha=0.05):
    """
    Performs a two-tailed Z-test, calculates the Z-statistic and P-value,
    and visualizes the results on a standard normal distribution plot.

    Args:
        sample_mean (float): The mean of the sample.
        population_mean (float): The known mean of the population (μ0).
        population_std (float): The known standard deviation of the population (σ).
        sample_size (int): The size of the sample (n).
        alpha (float, optional): The significance level (default is 0.05) for a two-tailed test.
    """
    # Calculate the standard error
    standard_error = population_std / np.sqrt(sample_size)

    # Calculate the Z-statistic
    z_statistic = (sample_mean - population_mean) / standard_error

    # Calculate the p-value (two-tailed)
    p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

    # Determine decision
    decision = "Reject H0" if p_value < alpha else "Fail to reject H0"

    print("--- Z-Test Results ---")
    print(f"Sample Mean: {sample_mean}")
    print(f"Hypothesized Population Mean (H0): {population_mean}")
    print(f"Known Population Std Dev: {population_std}")
    print(f"Sample Size: {sample_size}")
    print(f"Significance Level (alpha): {alpha}")
    print(f"Calculated Z-statistic: {z_statistic:.4f}")
    print(f"Calculated P-value: {p_value:.4f}")
    print(f"Decision: {decision}")

    # --- Visualization ---

    # Generate x values for the standard normal distribution
    x_vals = np.linspace(-4, 4, 1000)
    # Calculate the PDF for the standard normal distribution
    pdf_vals = norm.pdf(x_vals)

    # Calculate the critical Z-values for a two-tailed test
    z_critical_upper = norm.ppf(1 - alpha / 2)
    z_critical_lower = norm.ppf(alpha / 2)

    plt.figure(figsize=(10, 6))
    plt.plot(x_vals, pdf_vals, label='Standard Normal Distribution')

    # Shade the rejection regions
    plt.fill_between(x_vals, 0, pdf_vals, where=(x_vals >= z_critical_upper), color='red', alpha=0.3, label=f'Rejection Region (α/2 = {alpha/2})')
    plt.fill_between(x_vals, 0, pdf_vals, where=(x_vals <= z_critical_lower), color='red', alpha=0.3)

    # Shade the acceptance region
    plt.fill_between(x_vals, 0, pdf_vals, where=((x_vals > z_critical_lower) & (x_vals < z_critical_upper)), color='green', alpha=0.2, label='Acceptance Region (1 - α)')

    # Plot the critical values
    plt.axvline(z_critical_upper, color='red', linestyle='--', linewidth=1.5)
    plt.axvline(z_critical_lower, color='red', linestyle='--', linewidth=1.5, label=f'Critical Z = ±{z_critical_upper:.2f}')

    # Plot the calculated Z-statistic
    plt.axvline(z_statistic, color='blue', linestyle='-', linewidth=2, label=f'Sample Z = {z_statistic:.2f}')
    plt.text(z_statistic, norm.pdf(z_statistic) * 0.8, f'{z_statistic:.2f}', color='blue', ha='center')


    plt.title(f'Two-Tailed Z-Test (H0: μ = {population_mean})')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()

# Example Usage for performing the test and visualizing the sample Z-score
plot_z_test_with_sample(
    sample_mean=32.0,
    population_mean=30.0,
    population_std=5.0,
    sample_size=40,
    alpha=0.05
)

# Another example where we might fail to reject H0
plot_z_test_with_sample(
    sample_mean=30.5,  # Sample mean closer to the hypothesized mean
    population_mean=30.0,
    population_std=5.0,
    sample_size=40,
    alpha=0.05
)
```

```python
#5. Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05):
    """
    Visualizes Type 1 and Type 2 errors for a one-sample Z-test.

    Parameters:
    - mu0: Mean under H0 (null hypothesis)
    - mu1: Mean under H1 (alternative hypothesis)
    - sigma: Population standard deviation
    - n: Sample size
    - alpha: Significance level
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value for right-tailed test (change for two-tailed if needed)
    z_crit = norm.ppf(1 - alpha)
    x_crit = mu0 + z_crit * se

    # Type 1 error region: Reject H0 when it is true
    # Type 2 error region: Fail to reject H0 when H1 is true
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)
    y0 = norm.pdf(x, mu0, se)
    y1 = norm.pdf(x, mu1, se)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='H0: μ = {:.2f}'.format(mu0))
    plt.plot(x, y1, label='H1: μ = {:.2f}'.format(mu1), linestyle='--')

    # Shade Type 1 error (alpha)
    plt.fill_between(x, 0, y0, where=(x > x_crit), color='red', alpha=0.4, label='Type 1 error (α)')

    # Shade Type 2 error (beta)
    plt.fill_between(x, 0, y1, where=(x < x_crit), color='blue', alpha=0.3, label='Type 2 error (β)')

    plt.axvline(x_crit, color='black', linestyle=':', label='Critical value')
    plt.title('Type 1 and Type 2 Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.show()

    # Calculate Type 2 error (beta)
    beta = norm.cdf((x_crit - mu1) / se)
    print(f"Type 1 error (α): {alpha:.3f}")
    print(f"Type 2 error (β): {beta:.3f}")

# Example usage:
visualize_type1_type2_errors(mu0=0, mu1=1, sigma=1, n=30, alpha=0.05)
```

```python
#6. Write a Python program to perform an independent T-test and interpret the results

import numpy as np
from scipy import stats

# Generate two independent samples
np.random.seed(42)
group1 = np.random.normal(loc=10.5, scale=1.0, size=21)  # Example: Factory 1
group2 = np.random.normal(loc=9.5, scale=1.0, size=25)   # Example: Factory 2

# Perform independent t-test
t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=True)  # Use equal_var=False for Welch's t-test

print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (the group means are significantly different).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference between group means).")
```

```python
#9. Write a Python function to calculate the confidence interval for a sample mean and explain its significance.

import numpy as np
from scipy import stats

def confidence_interval_mean(sample, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        sample (array-like): The sample data.
        confidence (float): Confidence level (default is 0.95 for 95%).

    Returns:
        (mean, lower_bound, upper_bound)
    """
    sample = np.array(sample)
    n = len(sample)
    mean = np.mean(sample)
    sem = stats.sem(sample)  # Standard error of the mean
    df = n - 1
    t_crit = stats.t.ppf((1 + confidence) / 2, df)
    margin = t_crit * sem
    lower = mean - margin
    upper = mean + margin
    return mean, lower, upper

# Example usage:
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=30)
mean, lower, upper = confidence_interval_mean(data, confidence=0.95)
print(f"Sample mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
#10. Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy import stats

def margin_of_error(sample, confidence=0.95):
    """
    Calculate the margin of error for a given confidence level using sample data.

    Parameters:
        sample (array-like): The sample data.
        confidence (float): The confidence level (default is 0.95 for 95%).

    Returns:
        margin (float): The margin of error.
    """
    sample = np.array(sample)
    n = len(sample)
    sample_std = np.std(sample, ddof=1)
    standard_error = sample_std / np.sqrt(n)
    df = n - 1
    t_crit = stats.t.ppf((1 + confidence) / 2, df)
    margin = t_crit * standard_error
    return margin

# Example usage:
np.random.seed(1)
data = np.random.normal(loc=50, scale=8, size=40)
moe = margin_of_error(data, confidence=0.95)
print(f"Margin of Error (95% confidence): {moe:.2f}")
```

```python
#11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.

def bayes_theorem(prior_A, prior_B, likelihood_B_given_A):
    """
    Calculate the posterior probability using Bayes' Theorem.

    Parameters:
        prior_A: P(A) - Prior probability of hypothesis A
        prior_B: P(B) - Probability of evidence B
        likelihood_B_given_A: P(B|A) - Likelihood of B given A

    Returns:
        Posterior probability P(A|B)
    """
    posterior = (prior_A * likelihood_B_given_A) / prior_B
    return posterior

# Given probabilities
p_rain = 0.20            # P(rain)
p_cloudy = 0.40          # P(cloudy)
p_cloudy_given_rain = 0.85 # P(cloudy | rain)

# Calculate P(rain | cloudy)
p_rain_given_cloudy = bayes_theorem(p_rain, p_cloudy, p_cloudy_given_rain)
print(f"P(rain | cloudy) = {p_rain_given_cloudy:.3f}")  # Output: 0.425
```

```python
#12. Perform a Chi-square test for independence between two categorical variables in Python.

import pandas as pd

# Example data: Gender vs. Preferred Morning Drink
data = {
    'Gender': ['Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],
    'Drink': ['Coffee', 'Celsius', 'Coffee', 'Orange Juice', 'Celsius', 'Coffee', 'Orange Juice', 'Celsius', 'Coffee', 'Orange Juice']
}

df = pd.DataFrame(data)
contingency_table = pd.crosstab(df['Gender'], df['Drink'])
print(contingency_table)

from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f"Chi-square Statistic: {chi2:.4f}")
print(f"P-value: {p:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies Table:")
print(expected)
```
